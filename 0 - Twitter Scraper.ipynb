```python
import twint
```

```python
# Configure
c = twint.Config()
#c.Username = "ajtowers"
c.Search = ['NFLX', 'Netflix']
c.Since  = "2018-07-31"
c.Until  = "2018-11-25"
#c.Custom["id"] = ["id"]
c.Custom["date"] = ["date"]
c.Custom["tweet"] = ["tweet"]
#c.Format = "{date}/{tweet}"
c.Store_csv = True
c.Output = 'tweets'
```

```python
# Run
twint.run.Search(c)
```
