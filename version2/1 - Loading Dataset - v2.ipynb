## Lectura del Dataset

```python
# archivo de lectura
dir_name   = '../data/original'
file_name  = 'nflx_all_tweets_raw.txt'
filename = os.path.join(dir_name, file_name)
```

```python
filename
```

```python
# defino regex para extraer fecha y urls-pics
match_date = re.compile(r"(\d{4}-\d{2}-\d{2})", re.IGNORECASE)
match_url = re.compile(r"((https?|http):\/\/(www\.)?[a-z0-9\.:].*?(?=\s))", re.IGNORECASE)
match_pic = re.compile(r"pic.twitter.com/[\w].*", re.IGNORECASE)
#match_arrobas  = re.compile(r'(@\w+)', re.IGNORECASE)
#match_hashtags = re.compile(r"(#\w*)", re.IGNORECASE)
#match_signs = re.compile(r'(&(#\d*)+;)', re.IGNORECASE)
#only_words = re.compile(r'[^a-zA-Z ]', re.IGNORECASE)
```

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer(strip_handles=True, reduce_len=True)

with open(filename,'r') as f:

    # inicializo dataframe
    tweets = pd.DataFrame()

    # lectura del file linea por linea (en caso de que el archivo sea muy grande)
    for line in f:

        # extraigo la fecha, '' si retorna None
        date = getattr(re.search(match_date, line), 'groups', lambda:[u""])()[0]
        
        # reemplazo signo $ y '\n' por ''
        texto = line.replace('$', '')
        texto = texto.replace('\n', '')        
        
        # elimino urls, pics, hashtags y arrobas
        texto = re.sub(match_url, '', texto)
        texto = re.sub(match_pic, '', texto)
        
        # aplico TweetTokenizer
        texto = tknzr.tokenize(texto)
        
        
        #texto = re.sub(match_arrobas, '', texto)
        #texto = re.sub(match_hashtags, '', texto)
        #texto = re.sub(match_signs, '', texto)
        
        # elimino los encabezados (nro de tweet, date, hora, cod y posteador)
        #texto = ' '.join(texto.split(' ')[5:])
        
        # me quedo son con las palabras
        #texto = re.sub(only_words, "", texto)
      
        if date != '':
            # creo diccionario para agregar a dataframe
            d = {
                'date' : date,
                'text' : texto,
            }
            tweets = tweets.append(d, ignore_index=True)
```

```python
tweets = tweets.sort_values(by='date')
```
